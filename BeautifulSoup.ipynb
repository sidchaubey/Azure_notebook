```
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.moneycontrol.com/")
print(response.status_code)
```

```
print(response.headers)

print(response.headers.get("Content-Type"))
```

```
data = BeautifulSoup(response.text, "html.parser")
data
```

```
link_list = data.find_all("a")
```

```
list_urls = []
try:
  for i in link_list:
    list_urls.append(i["href"])
except Exception as e:
  print(e)
```

```
headers = response.headers
```

```
list_urls
page_responses = []
for i in list_urls:
  if i.startswith("https",0, len("https")):
    response = requests.get(i, allow_redirects=False)
    page_responses.append(response.text)
```

```
image_url = []
for i in page_responses:
  data = BeautifulSoup(i, "html.parser")
  link_list = data.find_all("img")
  image_url.append(link_list)
  print(data.prettify())
```

```
image_url
```
